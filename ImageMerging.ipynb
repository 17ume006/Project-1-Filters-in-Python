```python
#this program in python depicts the use of different modules in python to merge two images.
import numpy as Number
import matplotlib.pyplot as math
from PIL import Image
```

```python
water = Image.open('water.jpg')
#print(water.mode)
```

```python
print('water(RGB)')
math.imshow(water)
math.show()

r,g,b = water.split()
print('red channel')
math.imshow(r)
math.show()

print('green channel')
math.imshow(g)
math.show()

print('blue channel')
math.imshow(b)
math.show()
```

```python
back = Image.open('back.jpg')
#print(back.mode)
print('back(rgb)')
math.imshow(back)
math.show()

rB,gB,bB = back.split()
print('red channel')
math.imshow(rB)
math.show()

print('green channel')
math.imshow(gB)
math.show()

print('blue channel')
math.imshow(bB)
math.show()
```

```python
nimage = Image.merge('RGB',(r,gB,bB))
math.imshow(nimage)
math.xticks([]),math.yticks([])
math.show()
```
